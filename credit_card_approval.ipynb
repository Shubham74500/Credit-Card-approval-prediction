```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
column_names = ["A1", "A2","A3","A4", "A5", "A6", "A7", "A8", "A9", "A10", "A11", "A12", "A13", "A14", "A15","A16"]
df=pd.read_csv("C:/Users/HP/Downloads/archive (3)/dataset/crxdata.csv",header=None,names=column_names)
display(df.head())
display(df.info())
```

```python
display(df.isna().sum())
```

```python
#no value is missing 
```

```python
df['A2']=pd.to_numeric(df['A2'],errors='coerce')
df=df.drop('A14',axis=1)
```

```python
df=df.replace("?",np.NaN)
for col in df:
    if df[col].dtypes=='object':
        df[col]=df[col].fillna(df[col].value_counts().index[0])
    else:
        df[col]=df[col].fillna(df[col].mean())
print(df.isnull().sum())
```

```python
df['A1'] = df['A1'].apply(lambda x: 0 if x=='a' else 1)
df['A9'] = df['A9'].apply(lambda x: 0 if x=='t' else 1)
df['A10'] = df['A10'].apply(lambda x: 0 if x=='t' else 1)
df['A12'] = df['A12'].apply(lambda x: 0 if x=='t' else 1)
df['A16'] = df['A16'].apply(lambda x: 0 if x=='+' else 1)
df = pd.get_dummies(df)
display(df)
```

```python
from sklearn.model_selection import train_test_split

Y=df['A16'].values
X=df.drop(['A16'],axis=1).values
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.33,random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler=MinMaxScaler()
rescaledX_train=scaler.fit_transform(X_train)
rescaledX_test=scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

logres=LogisticRegression()

logres.fit(rescaledX_train,Y_train)
```

```python
from sklearn.metrics import confusion_matrix

y_pred=logres.predict(rescaledX_test)

print("Accuracy of the current model is :",logres.score(rescaledX_test,Y_test))

print("confusion matrix :\n",confusion_matrix(Y_test,y_pred))
```

```python
from sklearn.model_selection import GridSearchCV

tol=[0.01,0.001,0.0001]
max_iter=[100,150,200]

param_grid=dict(tol=tol,max_iter=max_iter)
```

```python
grid_model=GridSearchCV(estimator=logres,param_grid=param_grid,cv=5)

grid_model_result=grid_model.fit(rescaledX_train,Y_train)

best_score,best_param=grid_model_result.best_score_,grid_model_result.best_params_

best_model = grid_model_result.best_estimator_
best_model.predict(rescaledX_test)

print("Accuracy of logistic regression classifier: ", best_model.score(rescaledX_test, Y_test))
```
